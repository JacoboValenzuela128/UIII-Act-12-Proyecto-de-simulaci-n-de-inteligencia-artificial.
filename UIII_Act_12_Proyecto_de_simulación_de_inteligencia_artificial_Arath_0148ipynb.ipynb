# AGRICULTOR

```python
# ==========================================================
# SIMULADOR IA: AGRO-IA PRO v6.0 (Programación Orientada a Objetos)
# ==========================================================

class Agricultor:
    # Atributo de clase: funciona como la base de datos global de la IA
    lista_agricultores = []

    # El constructor de la clase (Molde con los atributos requeridos)
    def __init__(self, cultivo, hectareas, maquinaria, temporada, ubicacion, fertilizante):
        self.tipo_cultivo = cultivo
        self.hectareas = hectareas
        self.maquinaria_agricola = maquinaria
        self.temporada_cosecha = temporada
        self.ubicacion_terreno = ubicacion
        self.fertilizantes_usados = fertilizante
        # Atributo calculado automáticamente por la IA
        self.diagnostico_ia = ""

    # --- FUNCIÓN INTERNA DE IA ---
    def generar_diagnostico_ia(self):
        """Simula el motor de decisión de la IA analizando el objeto actual"""
        if self.tipo_cultivo.lower() == "maiz" and self.fertilizantes_usados.lower() == "quimico":
            self.diagnostico_ia = "⚠️ Riesgo de erosión"
        elif self.fertilizantes_usados.lower() == "organico" and self.hectareas <= 10:
            self.diagnostico_ia = "🌱 Eco-Sustentable"
        elif self.hectareas > 50:
            self.diagnostico_ia = "🚜 Alerta de Escala"
        else:
            self.diagnostico_ia = "✅ Rendimiento Óptimo"

    # --- CRUD: FUNCIONES DE GESTIÓN (Dentro de la clase) ---

    def agregar_agricultor(self):
        """Ejecuta la IA para el objeto actual y lo guarda en la lista"""
        self.generar_diagnostico_ia()
        # Accedemos a la lista de la clase para guardar este objeto (self)
        Agricultor.lista_agricultores.append(self)
        print(f"\n🤖 AGRO-IA: ¡Terreno registrado con éxito!")
        print(f"Diagnóstico de la IA: {self.diagnostico_ia}")

    @classmethod
    def listar_agricultores(cls):
        """Muestra todos los objetos guardados en forma de tabla bien alineada"""
        if cls.lista_agricultores == []:
            print("\n❌ La base de datos está vacía. No hay agricultores.")
        else:
            print("\n" + "="*120)
            print("                                    REPORTE GLOBAL GENERADO POR AGRO-IA O.O.P.                                  ")
            print("" + "="*120)
            # Formato de columnas perfectamente espaciadas
            print("{:<5} {:<15} {:<12} {:<15} {:<15} {:<15} {:<15} {:<20}".format(
                "ID", "Cultivo", "Hectáreas", "Maquinaria", "Temporada", "Ubicación", "Fertilizante", "Diagnóstico IA"
            ))
            print("-" * 120)

            # Ciclo FOR para recorrer la lista de OBJETOS
            contador = 1
            for agri in cls.lista_agricultores:
                print("{:<5} {:<15} {:<12} {:<15} {:<15} {:<15} {:<15} {:<20}".format(
                    contador,
                    agri.tipo_cultivo,
                    str(agri.hectareas) + " ha",
                    agri.maquinaria_agricola,
                    agri.temporada_cosecha,
                    agri.ubicacion_terreno,
                    agri.fertilizantes_usados,
                    agri.diagnostico_ia
                ))
                contador = contador + 1
            print("="*120)

    @classmethod
    def actualizar_agricultor(cls, id_registro):
        """Busca un objeto por su ID y permite modificar sus atributos"""
        # Ajustamos el ID ya que visualmente empieza en 1 pero en programación en 0
        indice = id_registro - 1

        # Validación básica de rango con IF sin usar try
        if indice >= 0 and indice < len(cls.lista_agricultores):
            # Obtenemos el objeto directamente de la lista
            agri_a_editar = cls.lista_agricultores[indice]

            print(f"\n--- Modificando Registro del Cultivo de: {agri_a_editar.tipo_cultivo} ---")
            agri_a_editar.tipo_cultivo = input("Nuevo tipo de cultivo: ")
            nueva_h = input("Nuevas hectáreas: ")
            agri_a_editar.hectareas = int(nueva_h)
            agri_a_editar.fertilizantes_usados = input("Nuevo tipo de fertilizante (quimico/organico): ")

            # Volvemos a calcular su diagnóstico con los nuevos datos
            agri_a_editar.generar_diagnostico_ia()
            print("\n🤖 AGRO-IA: ¡Registro actualizado y re-evaluado!")
        else:
            print("❌ ID no encontrado en el sistema.")

    @classmethod
    def borrar_agricultor(cls, id_registro):
        """Elimina un objeto de la lista mediante su ID"""
        indice = id_registro - 1
        if indice >= 0 and indice < len(cls.lista_agricultores):
            # Eliminamos el objeto usando el método pop de las listas
            eliminado = cls.lista_agricultores.pop(indice)
            print(f"\n♻️ Sistema: Se eliminó el cultivo de {eliminado.tipo_cultivo} exitosamente.")
        else:
            print("❌ ID inválido. No se borró nada.")


# ==========================================================
# MENÚ INTERACTIVO (CONTROL PRINCIPAL DEL SISTEMA)
# ==========================================================
while True:
    print("\n--- MENÚ PRINCIPAL AGRO-IA (MODO POO) ---")
    print("1. Agregar Agricultor (Crear Objeto)")
    print("2. Listar Agricultores (Ver Tabla)")
    print("3. Actualizar Agricultor")
    print("4. Borrar Agricultor")
    print("5. Salir")

    opcion = input("\nSelecciona una opción (1-5): ")

    # 1. AGREGAR
    if opcion == "1":
        print("\n--- CAPTURA DE DATOS (ATRIBUTOS DE OBJETO) ---")
        c = input("Tipo de cultivo: ")
        h = input("Hectáreas: ")
        h = int(h)
        m = input("Maquinaria agrícola: ")
        t = input("Temporada de cosecha: ")
        u = input("Ubicación del terreno: ")
        f = input("Fertilizantes usados (quimico / organico): ")

        # CREACIÓN DEL OBJETO: Instanciamos la clase con sus datos
        nuevo_agricultor = Agricultor(c, h, m, t, u, f)
        # LLAMADA AL MÉTODO: El objeto se guarda a sí mismo
        nuevo_agricultor.agregar_agricultor()

    # 2. LISTAR
    elif opcion == "2":
        # Llamamos al método directamente desde la Clase (método de clase)
        # No requerimos un objeto específico para ver la tabla completa
        Agricultor.listar_agricultores()

    # 3. ACTUALIZAR
    elif opcion == "3":
        # Primero listamos para que el usuario sepa qué ID elegir
        Agricultor.listar_agricultores()
        if Agricultor.lista_agricultores != []:
            id_modificar = input("\nIntroduce el ID del registro que deseas actualizar: ")
            id_modificar = int(id_modificar)
            Agricultor.actualizar_agricultor(id_modificar)

    # 4. BORRAR
    elif opcion == "4":
        # Primero listamos para ver los ID disponibles
        Agricultor.listar_agricultores()
        if Agricultor.lista_agricultores != []:
            id_borrar = input("\nIntroduce el ID del registro que deseas eliminar: ")
            id_borrar = int(id_borrar)
            Agricultor.borrar_agricultor(id_borrar)

    # 5. SALIR
    elif opcion == "5":
        print("\nCerrando sistema de Inteligencia Artificial Agrícola Orientada a Objetos...")
        break

    else:
        print("❌ Opción incorrecta. Elige de la lista.")

    print("\n" + "-"*50)
```
